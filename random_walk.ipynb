```python
from random import choice

class RandomWalk():
    """A class to generate random walks."""
    
    def __init__(self, num=10000):
        """Initialize attributes of the walk."""
        self.num = num
        
        # Start at point (0,0)
        self.x_vals = [0]
        self.y_vals = [0]
        
    def fill_walk(self):
        """Calculate all the points in the walk."""

        while len(self.x_vals) < self.num:
            x_dir = choice([-1, 1])
            x_dist = choice([0, 1, 2, 3, 4])
            x_step = x_dir * x_dist

            y_dir = choice([-1, 1])
            y_dist = choice([0, 1, 2, 3, 4])
            y_step = y_dir * y_dist

            # Ignore if there is no movement.
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the next (x,y) values.
            next_x = self.x_vals[-1] + x_step
            next_y = self.y_vals[-1] + y_step

            self.x_vals.append(next_x)
            self.y_vals.append(next_y)
```

```python
rw = RandomWalk()
rw.fill_walk()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(rw.x_vals, rw.y_vals, edgecolor='none', s=1.5)
plt.show()
```
